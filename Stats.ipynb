### Estudos de Estatística

```python
import random
import numpy as np
import matplotlib.pyplot as pyplot
import scipy.stats as stats
import time
```

```python
def rollDie(times):    
  die_rolls = []
  for i in range(times):
     die_rolls.append(random.randint(1,6)) 
  # retorna os dados
  return die_rolls    
```

## Roll the die

```python
# lança os dados 10 vezes
rolls = rollDie(10)
```

```python
print(rolls)
```

### Plota os lançamentos

```python
for t in range(4):   
  rolls = rollDie(1000)    
  pyplot.hist(rolls)
  pyplot.title('Lançamentos do dado')
  pyplot.show()
  pyplot.close()
  time.sleep(1)  
```

```python
# média dos lançamentos do dado
print('média ->', np.mean(rolls))

# desvio padrão
print('Desvio padrão->', np.std(rolls))
```

### Central Limit Theorem

```python
## pega uma amostra dos meus dados
```

```python
averages = []

## pega várias amostras
for x in range(1000):
  sample = []    
  # extrai a amostra
  for i in range(4):
    sample.append(random.choice(rolls))
    
  # pega a média da amostra que tiramos e guarda-a
  averages.append(np.mean(sample))  

# mostra a distribuição da amostra
pyplot.hist(sample)
pyplot.title('Sample')
pyplot.show()  


# mostra a distribuição das médias das amostras
pyplot.hist(averages, bins=40)
pyplot.title('Médias das amostras - CLT')
pyplot.show()  



```
